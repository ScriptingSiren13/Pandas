```python
import pandas as pd
```

```python
people={
    "first":["Isa","Jane","John"],
    "last":["Bell","Doe","Brown"],
    "email":["isabell@xyz.com","janedoe@xyz.com","johnbrown@xyz.com"]
}
```

```python
df=pd.DataFrame(people)
df
```

```python
df['email']
```

#### To set a column as an index

```python
df.set_index('email')
#the changes will not be permanent in the df unless we specifically tell it to do so 
```

#### To set a column as an index permanently

```python
df.set_index('email',inplace=True)
df
```

```python
#Tells us the index of our data
df.index
```

```python
#Its usage:
df.loc['isabell@xyz.com']
```

```python
#using it for a particular entry:
df.loc['isabell@xyz.com','last']
```

```python
#If we use an integer inside the loc, it will give an error because it no longer has an index with that lablel
df.loc[0]
```

```python
#Thus we can use iloc for integer index:
df.iloc[0]
```

#### To reset an index

```python
df.reset_index(inplace=True)
df
```

```python
df.index #email no longer an index
```
